```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') # For plots
import itertools
from sklearn.metrics import mean_squared_error
```

```python
path = '../data/'
weather_true = pd.read_csv(path + 'HistoricalWeather.csv')
weather_pred = pd.read_csv(path + 'HistoricalWeatherForecasts.csv')
```

```python
def compare_pred(true, pred, lead_time = 5):
    
    # prepare dfs
    col_map = {
        'Date': 'date',
        'Region': 'region',
        'Parameter': 'parameter',
        'count()[unit: km^2]': 'count',
        'min()': 'min',
        'max()': 'max',
        'mean()': 'mean',
        'variance()': 'variance',
        'Lead time': 'lead_time',
    }
    true = true.copy().rename(columns = col_map).set_index('date')
    pred = pred.copy().rename(columns = col_map).set_index('date')
    true.index = pd.to_datetime(true.index)
    pred.index = pd.to_datetime(pred.index)
    
    # filter lead_time
    pred = pred[pred['lead_time'] == lead_time].drop('lead_time', axis = 1)
    
    # combination of regions and parameters
    result = []
    regions = np.unique(pred['region'])
    parameters = np.unique(pred['parameter'])
    for region, parameter in itertools.product(regions, parameters):
        df_true = true[(true['region'] == region) & (true['parameter'] == parameter)]
        df_pred = pred[(pred['region'] == region) & (pred['parameter'] == parameter)]
        df = df_true.join(df_pred, how = 'inner', lsuffix = '_true', rsuffix = '_pred').dropna()
        result.append({
            'region': region,
            'parameter': parameter,
            'lead_time': lead_time,
            'corr(min)': df['min_true'].corr(df['min_pred']), 
            'corr(max)': df['max_true'].corr(df['max_pred']), 
            'corr(mean)': df['mean_true'].corr(df['mean_pred']),
            'rmse(min)': np.sqrt(mean_squared_error(df['min_true'], df['min_pred'])), 
            'rmse(max)': np.sqrt(mean_squared_error(df['max_true'], df['max_pred'])), 
            'rmse(mean)': np.sqrt(mean_squared_error(df['mean_true'], df['mean_pred'])),
        })
    
    return pd.DataFrame(result)
```

```python
corr = []
rmse = []
for lead_time in [5, 10, 15]:
    lead = compare_pred(weather_true, weather_pred, lead_time = lead_time)
    corr.append(lead.melt(id_vars = ['region', 'parameter', 'lead_time'], 
                          value_vars = ['corr(min)', 'corr(max)', 'corr(mean)']))
    rmse.append(lead.melt(id_vars = ['region', 'parameter', 'lead_time'], 
                          value_vars = ['rmse(min)', 'rmse(max)', 'rmse(mean)']))
```

```python
corr_res = pd.concat(corr)
corr_res
```

```python
rmse_res = pd.concat(rmse)
rmse_res
```

```python
plt.title('Lead time 5 days correlation')
sns.scatterplot(x = 'parameter', y = 'value', hue = 'variable', data = corr_res[corr_res['lead_time'] == 5])
plt.show()
```
