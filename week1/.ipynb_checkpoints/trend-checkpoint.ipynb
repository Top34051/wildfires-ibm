```python
import numpy as np
import pandas as pd
import datetime
import dateutil

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style("whitegrid")
```

```python
# fill na with median
def fill_na_median(ori):
    df = ori.copy()
    
    std_med = np.median(df["Std_confidence"].dropna())
    df["Std_confidence"] = df["Std_confidence"].fillna(std_med)

    var_med = np.median(df["Var_confidence"].dropna())
    df["Var_confidence"] = df["Var_confidence"].fillna(std_med)
    
    return df
```

```python
def clean(ori):
    df = ori.copy()
    df["Date"] = [dateutil.parser.parse(date) for date in df["Date"]]
    df = fill_na_median(df)
    return df
```

```python
# split into region
def split_region(df):
    regions = np.unique(df["Region"])

    df_region = {}
    for region in regions:
        df_region[region] = df[df["Region"] == region]
    
    return regions, df_region
```

```python
fire = pd.read_csv("Nov_10/Historical_Wildfires.csv")
fire = clean(fire)

fire.info()
```

```python
regions, df_region = split_region(fire)
```

```python
for region, df in df_region.items():
#     df_2010 = df[("1-1-2010" <= df["Date"]) & (df["Date"] < "1-1-2011")]
    
    plt.figure(figsize = (16, 6))
    sns.lineplot(x = "Date", y = "Estimated_fire_area", data = df)
    plt.title(region)
```

```python
def period(df, days):
    parameters = ["Estimated_fire_area", 
                  "Mean_estimated_fire_brightness", 
                  "Mean_estimated_fire_radiative_power"]
    df = df.pivot_table(values = parameters, 
                        index = "Date",
                        columns = "Region")
    df = df.reset_index().fillna(0)
    return df.groupby(pd.Grouper(key = "Date", freq = f"{days*24}h")).mean()
```

```python
df_30 = period(fire, 30)

plt.figure(figsize = (20, 12))
sns.heatmap(df_30.corr(), cmap = "coolwarm", xticklabels = True, yticklabels = True)
```

```python
def filter(ori, region = None, start_date = "01-01-2005", end_date = "01-01-2022"):
    df = ori.copy()
    df = df[(start_date <= df["Date"]) & (df["Date"] <= end_date)]
    if region != None:
        if isinstance(region, list):
            df = df[df["Region"].isin(region)]
        else:
            df = df[df["Region"] == region]
    return df

def plot_yearly(df, columns = None):
    df = df.groupby("Date").median()

    idx = pd.MultiIndex.from_arrays([
            pd.to_datetime(df.index.strftime('2012-%m-%d %H:%M:%S')),
            df.index.year])

    d1 = df.set_index(idx).unstack().resample("W").mean()
        
    if columns == None:
        columns = d1.columns
       
    fig, axes = plt.subplots(len(columns), 1, figsize=(15, len(columns) * 3))
    idx = 0
    for col in columns:
        d1[col].plot(ax = axes[idx], title = col)
        legend = axes[idx].legend()
        legend.remove()
        idx += 1
        
    plt.show()
    
def plot_monthly(df, columns = None):
    df = df.groupby("Date").median()
    
    df["Day"] = pd.to_datetime(df.index.strftime('2012-1-%d %H:%M:%S'))
    df["Month"] = [str(date.month) for date in df.index]
    
    df = df.groupby(["Month", "Day"]).median()
    df = df.reset_index()
    
    return df
    
    if columns == None:
        columns = df.columns
        
#     sns.lineplot(x = "Day", y = "Estimated_fire_area", hue = "Month", data = df)
#     pass
        
    fig, axes = plt.subplots(len(columns), 1, figsize=(15, len(columns) * 10))
    idx = 0
    for col in columns:
        sns.lineplot(x = "Day", y = col, hue = "Month", ax = axes[idx], data = df)
        d1[col].plot(ax = axes[idx], title = col)
        legend = axes[idx].legend()
        legend.remove()
        idx += 1
        
    plt.show()
    
        
east = ["NSW", "TA", "VI"]
west = ["NT", "QL", "SA", "WA"]
```

```python
wants = ["Estimated_fire_area", "Mean_estimated_fire_brightness", "Mean_estimated_fire_radiative_power"]

plot_yearly(filter(fire, region = east), columns = wants)
```

```python
wants = ["Estimated_fire_area", "Mean_estimated_fire_brightness", "Mean_estimated_fire_radiative_power"]

plot_yearly(filter(fire, region = west), columns = wants)
```

```python
wants = ["Estimated_fire_area", "Mean_estimated_fire_brightness", "Mean_estimated_fire_radiative_power"]

plot_monthly(fire, columns = wants)
```
